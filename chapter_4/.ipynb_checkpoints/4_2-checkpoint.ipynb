```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


print("\n" + "=" * 60)
print("Problem 32: Random Matrix Determinants")
print("=" * 60)

print("\nTypical determinants of rand(n) and randn(n):")
print("-" * 60)

sizes = [50, 100, 200, 400]
num_trials = 10

for n in sizes:
    print(f"\nn = {n}:")
    
    # rand(n): uniform distribution [0, 1]
    rand_dets = [np.linalg.det(np.random.rand(n, n)) for _ in range(num_trials)]
    
    # randn(n): standard normal distribution
    randn_dets = [np.linalg.det(np.random.randn(n, n)) for _ in range(num_trials)]
    
    print(f"  rand({n}) determinants:")
    print(f"    Mean: {np.mean(rand_dets):.6e}")
    print(f"    Std:  {np.std(rand_dets):.6e}")
    print(f"    Range: [{np.min(rand_dets):.6e}, {np.max(rand_dets):.6e}]")
    
    print(f"  randn({n}) determinants:")
    print(f"    Mean: {np.mean(randn_dets):.6e}")
    print(f"    Std:  {np.std(randn_dets):.6e}")
    print(f"    Range: [{np.min(randn_dets):.6e}, {np.max(randn_dets):.6e}]")

print("\n\nWhat does 'Inf' mean in MATLAB?")
print("-" * 60)
print("'Inf' in MATLAB means infinity (positive infinity).")
print("It occurs when a calculation results in overflow (number too large)")
print("or when dividing by zero (e.g., 1/0 = Inf).")
print("In Python/NumPy, this is represented as 'inf' or 'np.inf'.")
print("\nExample:")
print(f"  1/0 in NumPy: {np.inf}")
print(f"  Very large number: {1e308 * 10}")

print("\n" + "=" * 60)
print("Problem 33: Largest Determinant of 4×4 Binary Matrix")
print("=" * 60)

from itertools import product

max_det = -np.inf
best_matrix = None

# Generate all possible 4x4 matrices with entries 0 or 1
# Total possibilities: 2^16 = 65,536
count = 0
for entries in product([0, 1], repeat=16):
    A = np.array(entries).reshape(4, 4)
    det_A = np.linalg.det(A)
    
    if det_A > max_det:
        max_det = det_A
        best_matrix = A.copy()
    
    count += 1

print(f"\nSearched through {count:,} possible matrices")
print(f"\nLargest determinant found: {max_det:.0f}")
print(f"\nMatrix that achieves this determinant:")
print(best_matrix.astype(int))

# Verify
print(f"\nVerification: det = {np.linalg.det(best_matrix):.6f}")

# Find how many matrices achieve this maximum
max_count = 0
for entries in product([0, 1], repeat=16):
    A = np.array(entries).reshape(4, 4)
    if abs(np.linalg.det(A) - max_det) < 1e-10:
        max_count += 1

print(f"\nNumber of matrices achieving max determinant: {max_count}")

print("\n" + "=" * 60)
print("All problems completed!")
print("=" * 60)
```

## The Hilbert matrix hilb(n) has i, j entry equal to 1/(i + j − 1). Print ti determinants of hilb(1),hilb(2),...,hilb(10). Hilbert matrices are hard to work with! What are the pivots?

```python

def hilbert_matrix(n):
    """Create Hilbert matrix of size n x n"""
    H = np.zeros((n, n))
    for i in range(1, n+1):
        for j in range(1, n+1):
            H[i-1, j-1] = 1 / (i + j - 1)
    return H
```

```python
hilbert_matrix(3)
```

```python
determinants = []
for n in range(1, 11):
    H = hilbert_matrix(n)
    #print(H)
    det_H = np.linalg.det(H)
    determinants.append(det_H)
    print(f"det(hilb({n:2d})) = {det_H:.6e}")
```

```python
for n in range(1, 11):
    H = hilbert_matrix(n)
    # LU decomposition with partial pivoting
    from scipy.linalg import lu
    P, L, U = lu(H)
    pivots = np.diag(U)
    print(f"\nhilb({n}) pivots:")
    print(pivots)
```

```python
# Visualize how determinants decay
plt.figure(figsize=(10, 6))
plt.semilogy(range(1, 11), np.abs(determinants), 'bo-', linewidth=2, markersize=8)
plt.xlabel('Matrix size n', fontsize=12)
plt.ylabel('|det(hilb(n))|', fontsize=12)
plt.title('Hilbert Matrix Determinants (log scale)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 11))
plt.tight_layout()
```
