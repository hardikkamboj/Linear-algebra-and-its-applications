```python
import numpy as np

# Define the matrix
I = np.eye(4)
ones = np.ones((4, 4))
A = 5 * I - ones

# Compute the inverse
A_inv = np.linalg.inv(A)

print("Matrix A:")
print(A)
print("\nInverse of A:")
print(A_inv)
```
