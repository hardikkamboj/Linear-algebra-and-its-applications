```python
import numpy as np
from scipy.linalg import solve, inv
import time





```

```python

# Create identity matrix and a random upper triangular matrix
I = np.eye(1000)
A = np.random.rand(1000, 1000)
B = np.triu(A)  # upper triangular matrix
```

```python
# Time for matrix inverse
start_inv = time.time()
B_inv = inv(B)
end_inv = time.time()
print("Time for inv(B):", end_inv - start_inv, "seconds")
```

```python
# Time for solving B x = I
start_solve = time.time()
B_solve = solve(B, I)  # preferred over np.linalg.solve for large/triangular systems
end_solve = time.time()
print("Time for B \\ I (solve):", end_solve - start_solve, "seconds")
```
